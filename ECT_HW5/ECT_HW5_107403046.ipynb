### 5

```python
lscrsum = 0.0
sscrsum = 0.0
lscrArr = []
sscrArr = []

for i in range(1,30):
    sample = df.groupby('Churn', group_keys=False).apply(lambda x: x.sample(frac=0.6))
    X = df
    y = df['Churn']

    X_train = sample.drop(['Churn'], axis = 1)
    y_train = sample['Churn']#要預測的欄位

    X_test = copy.deepcopy(df[~df.index.isin(sample.index.values)])
    y_test = df['Churn'].drop(sample['Churn'].index)
    
    print("第",i,"次：")
    
    
    #logistic regression
    print("logistic regression:")
    lmodel = LogisticRegression()
    lmodel.fit(X_train, y_train)

    lscores = cross_val_score(lmodel,X_test,y_test,cv=10,scoring='accuracy')
    print(lscores.mean())
    lscrsum+=lscores.mean()
    lscrArr.append(lscores.mean())
    
    
    #SVM
    print("SVM:")
    svcModel = svm.LinearSVC(C=1) 
    svcModel.fit(X_train, y_train)

    cscores = cross_val_score(svcModel,X_test,y_test,cv=10,scoring='accuracy')
    print(cscores.mean())
    sscrsum+=cscores.mean()
    sscrArr.append(cscores.mean())
    
    #Model=svm.SVC(kernel='linear',C=1)
    #Model.fit(X_train, y_train)

    #Cscores = cross_val_score(Model,X_test,y_test,cv=10,scoring='accuracy')
    #print(Cscores.mean())
    #sscrsum+=Cscores.mean()
    
print("logistic regression最終的平均accuracy:",lscrsum/30)
print("SVM最終的平均accuracy:",sscrsum/30)
```

### 6

```python
import scipy.stats as stats

stats.ttest_rel(lscrArr, sscrArr)
```

The test statistic comes out to be equal to 9.975 and the corresponding two-sided p-value is 1.017e-10.


The paired samples t-test follows the null and alternative hypotheses:

H0: 表示兩個model的Accuracy相等<br>
HA: 表示兩個model的Accuracy不相等

As the p-value comes out to be equal to 1.017e-10 which is less than 0.05 hence we reject the null hypothesis. So, we have enough proof to claim that the accuracy score of two models is different.
