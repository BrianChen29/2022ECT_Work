### 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('insurance.csv')
df
```

### 2

```python
#sex部分
df.loc[df.sex == 'male', 'sex'] = 0
df.loc[df.sex == 'female', 'sex'] = 1

#smoker部分 no = 0; yes = 1
df.loc[df.smoker == 'no', 'smoker'] = 0
df.loc[df.smoker == 'yes', 'smoker'] = 1

#region部分 S = 0; C = 1; Q = 2; 
df.loc[df.region == 'northeast', 'region'] = 0
df.loc[df.region == 'southeast', 'region'] = 1
df.loc[df.region == 'southwest', 'region'] = 2
df.loc[df.region == 'northwest', 'region'] = 3

df
```

### 3

```python
from sklearn import tree
from sklearn.linear_model import LinearRegression
from sklearn import svm
from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import export_graphviz
from sklearn.metrics import accuracy_score


X = df.drop(['charges'], axis = 1)
y = df['charges']#要預測的欄位

X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.5, random_state = 12)
```

### 4
#### linear regression部分

```python
#使用linear regresion演算法
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Obtain the coefficient of determination by calling the model with the score() function, then print the coefficient:
r_sq = regressor.score(X_test, y_test)
print('coefficient of determination:', r_sq)

# Print the Intercept:
w_0 = regressor.intercept_
print('intercept:', w_0)

# Print the Slope:
w_1 = regressor.coef_
print('Coefficient:', w_1) 
```

```python
# Predict a Response and print it:
yr_pred = regressor.predict(X_test)
print('Predicted response:', yr_pred, sep='\n')
```

### 4
#### SVM部分

```python
#kernel = linear
svmModel = svm.SVR(C=1,kernel='linear') #kernel : linear, poly, rbf
svmModel.fit(X_train, y_train)
```

```python
# Obtain the coefficient of determination by calling the model with the score() function, then print the coefficient:
svmscore = svmModel.score(X_test, y_test)
print('coefficient of determination:', svmscore)
```

```python
ys_pred = svmModel.predict(X_test)
print('Predicted response:', ys_pred, sep='\n')
```

```python
#kernel = poly
svmpoly = svm.SVR(C=1, kernel='poly', degree=3, gamma='auto')
svmpoly.fit(X_train, y_train)
```

```python
# Obtain the coefficient of determination by calling the model with the score() function, then print the coefficient:
polyscore = svmpoly.score(X_test, y_test)
print('coefficient of determination:', polyscore)
```

```python
ysp_pred = svmpoly.predict(X_test)
print('Predicted response:', ysp_pred, sep='\n')
```

```python
#kernel = rbf
svmrbf = svm.SVR(C=6, kernel='rbf', degree=3, gamma='auto')
svmrbf.fit(X_train, y_train)
```

```python
# Obtain the coefficient of determination by calling the model with the score() function, then print the coefficient:
rbfscore = svmrbf.score(X_test, y_test)
print('coefficient of determination:', rbfscore)
```

```python
ysr_pred = svmrbf.predict(X_test)
print('Predicted response:', ysr_pred, sep='\n')
```

### 4
#### Decision Tree

```python
#Decision tree無法處理continuous的型態，所以轉成整數型態
df.charges = df.charges.astype('int64')

X = df.drop(['charges'], axis = 1)
y = df['charges']#要預測的欄位

X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.5, random_state = 12)
```

```python
#使用決策樹演算法
dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
```

```python
# Predict a Response and print it:
ydt_pred = dtree.predict(X_test)
print('Predicted response:', yr_pred, sep='\n')
```

```python
# 衡量線性迴歸的R2、RMSE、MAE
from math import sqrt
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
print("R2 score(coefficient of determination):\nregression:",r2_score(y_test,yr_pred),"\nSVM:",r2_score(y_test,ysp_pred),"\nDecision tree:",r2_score(y_test,ydt_pred),"\n")
print("MAE(mean_absolute_error):\nregression:",mean_absolute_error(y_test,yr_pred),"\nSVM:",mean_absolute_error(y_test,ysp_pred),"\nDecision tree:",mean_absolute_error(y_test,ydt_pred),"\n")
print("RMSE:\nregression:",sqrt(mean_squared_error(y_test,yr_pred)),"\nSVM:",sqrt(mean_squared_error(y_test,ysp_pred)),"\nDecision tree:",sqrt(mean_squared_error(y_test,yr_pred)))
```
