```python
#將字串轉換成數字分類

#PreferredLoginDevice部分
df.loc[df.PreferredLoginDevice == 'Computer', 'PreferredLoginDevice'] = 0
df.loc[df.PreferredLoginDevice == 'Mobile Phone', 'PreferredLoginDevice'] = 1
df.loc[df.PreferredLoginDevice == 'Phone', 'PreferredLoginDevice'] = 2

#PreferredPaymentMode部分
df.loc[df.PreferredPaymentMode == 'CC', 'PreferredPaymentMode'] = 0
df.loc[df.PreferredPaymentMode == 'COD', 'PreferredPaymentMode'] = 1
df.loc[df.PreferredPaymentMode == 'Cash on Delivery', 'PreferredPaymentMode'] = 2
df.loc[df.PreferredPaymentMode == 'Credit Card', 'PreferredPaymentMode'] = 3
df.loc[df.PreferredPaymentMode == 'Debit Card', 'PreferredPaymentMode'] = 4
df.loc[df.PreferredPaymentMode == 'E wallet', 'PreferredPaymentMode'] = 5
df.loc[df.PreferredPaymentMode == 'UPI', 'PreferredPaymentMode'] = 6

#Gender部分
df.loc[df.Gender == 'Male', 'Gender'] = 0
df.loc[df.Gender == 'Female', 'Gender'] = 1

#PreferedOrderCat部分
df.loc[df.PreferedOrderCat == 'Fashion', 'PreferedOrderCat'] = 0
df.loc[df.PreferedOrderCat == 'Grocery', 'PreferedOrderCat'] = 1
df.loc[df.PreferedOrderCat == 'Laptop & Accessory', 'PreferedOrderCat'] = 2
df.loc[df.PreferedOrderCat == 'Mobile', 'PreferedOrderCat'] = 3
df.loc[df.PreferedOrderCat == 'Mobile Phone', 'PreferedOrderCat'] = 4
df.loc[df.PreferedOrderCat == 'Others', 'PreferedOrderCat'] = 5

#MaritalStatus部分
df.loc[df.MaritalStatus == 'Divorced', 'MaritalStatus'] = 0
df.loc[df.MaritalStatus == 'Married', 'MaritalStatus'] = 1
df.loc[df.MaritalStatus == 'Single', 'MaritalStatus'] = 2

df
```

```python
#空值部分直接以平均數帶入
df.fillna(df.mean(),inplace=True)
df
```

### 5

```python
from sklearn import tree
from sklearn.linear_model import LinearRegression
from sklearn import svm
from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import export_graphviz
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score


X = df.drop(['Churn'], axis = 1)
y = df['Churn']#要預測的欄位

X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.5, random_state = 12)
```

### 6,7
SVM部分

```python
#kernel = linear
svmModel = svm.SVR(C=1,kernel='linear') 
svmModel.fit(X_train, y_train)
```

```python
# Obtain the coefficient of determination by calling the model with the score() function, then print the coefficient:
svmscore = svmModel.score(X_test, y_test)
print('coefficient of determination:', svmscore)
```

```python
ys_pred = svmModel.predict(X_test)
print('Predicted response:', ys_pred, sep='\n')
```

```python
#### kernel = rbf
rbfModel = svm.SVR(C=6,kernel='rbf',degree=3,gamma='auto') 
rbfModel.fit(X_train, y_train)
```

```python
# Obtain the coefficient of determination by calling the model with the score() function, then print the coefficient:
rbfscore = rbfModel.score(X_test, y_test)
print('coefficient of determination:', rbfscore)
```

```python
ysr_pred = rbfModel.predict(X_test)
print('Predicted response:', ysr_pred, sep='\n')
```

```python
svcModel = svm.LinearSVC(C=1, max_iter=10000) 
svcModel.fit(X_train, y_train)
```

```python
# Obtain the coefficient of determination by calling the model with the score() function, then print the coefficient:
cscore = svcModel.score(X_test, y_test)
print('coefficient of determination:', cscore)
```

```python
ysvc_pred = svcModel.predict(X_test)
print('Predicted response:', ysvc_pred, sep='\n')
```

```python
crbfModel=svm.SVC(kernel='rbf', gamma=0.7, C=1)
crbfModel.fit(X_train, y_train)
```

```python
# Obtain the coefficient of determination by calling the model with the score() function, then print the coefficient:
crscore = crbfModel.score(X_test, y_test)
print('coefficient of determination:', crscore)
```

```python
Model=svm.SVC(kernel='linear',C=1)
Model.fit(X_train, y_train)
```

```python
# Obtain the coefficient of determination by calling the model with the score() function, then print the coefficient:
score = Model.score(X_test, y_test)
print('coefficient of determination:', score)
```

最後選擇此SVM模型 accuracy最高

```python
y_pred = Model.predict(X_test)
print('Predicted response:', y_pred, sep='\n')
```

### 8

```python
svcaccuracy = accuracy_score(y_test, y_pred)
lraccuracy = accuracy_score(y_test, yl_pred)
dtaccuracy = accuracy_score(y_test, ydt_pred)

print("SVM accuracy: ", svcaccuracy,"\nLogistic Regression accuracy: ", lraccuracy,"\nDecision Tree accuracy:", dtaccuracy)
```
