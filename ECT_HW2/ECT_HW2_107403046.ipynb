### 2.3

```python
import numpy as np
m_age_ave = 0.0
f_age_ave = 0.0

#è¨ˆç®—åŽ»æŽ‰ç©ºå€¼å¾Œå‰©ä¸‹çš„ç”·ç”Ÿå¹´é½¡å¹³å‡æ•¸ï¼ˆå››æ¨äº”å…¥å–åˆ°å°æ•¸é»žç¬¬ä¸€ä½ï¼‰
male = df[df['Sex'] == 'male']
male = male.dropna(subset = ['Age'])
m_age_ave = male['Age'].sum()/len(male)
print(round(m_age_ave, 1))

#è¨ˆç®—åŽ»æŽ‰ç©ºå€¼å¾Œå‰©ä¸‹çš„å¥³ç”Ÿå¹´é½¡å¹³å‡æ•¸ï¼ˆå››æ¨äº”å…¥å–åˆ°å°æ•¸é»žç¬¬ä¸€ä½ï¼‰
female = df[df['Sex'] == 'female']
female = female.dropna(subset = ['Age'])
f_age_ave = male['Age'].sum()/len(female)
print(round(f_age_ave, 1))

df.loc[(df.Sex == 'male')&(df.Age.isnull()), 'Age'] = round(m_age_ave, 1)
df.loc[(df.Sex == 'female')&(df.Age.isnull()), 'Age'] = round(f_age_ave, 1)
df
```

### 2.4

```python
df['Cabin'] = df['Pclass'] * df['Fare']
df
```

### 2.5

```python
df.loc[df.Survived == 0, 'Survived'] = 'Alive'
df.loc[df.Survived == 1, 'Survived'] = 'Dead'
df
```

### 2.6

```python
#Sexéƒ¨åˆ†
df.loc[df.Sex == 'male', 'Sex'] = 0
df.loc[df.Sex == 'female', 'Sex'] = 1

#Embarkedéƒ¨åˆ† S = 0; C = 1; Q = 2 ç©ºå€¼=3ï¼ˆåœ¨å¾Œé¢è¨“ç·´æ™‚ç™¼ç¾æœ‰ç©ºå€¼å°Žè‡´éŒ¯èª¤ï¼‰
df.loc[df.Embarked == 'S', 'Embarked'] = 0
df.loc[df.Embarked == 'C', 'Embarked'] = 1
df.loc[df.Embarked == 'Q', 'Embarked'] = 2
df['Embarked']=df['Embarked'].fillna(3)

df
```

### 2.7

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import export_graphviz
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

X = df.drop(['PassengerId', 'Survived', 'Name', 'Ticket', 'Pclass'], axis=1)#æŠŠä¸éœ€è¦ä½œç‚ºè¨“ç·´è³‡æ–™çš„æ¬„ä½åŽ»æŽ‰ï¼Œå‰©ä¸‹çš„ç‚ºè¨“ç·´è³‡æ–™
y = df['Survived']#è¦é æ¸¬çš„æ¬„ä½

X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.5, random_state = 12, stratify=y) #å»ºç«‹ç‰¹å¾µXï¼Œèˆ‡ç›®æ¨™yï¼Œä¸¦å°‡è³‡æ–™å€åˆ†æˆè¨“ç·´é›†èˆ‡æ¸¬è©¦é›†ï¼Œæ¯”ä¾‹ç‚º50%
```

```python
#ä½¿ç”¨æ±ºç­–æ¨¹æ¼”ç®—æ³•
model = DecisionTreeClassifier(criterion='gini', max_depth=4)
dtree = model.fit(X_train,y_train)
dtree
```

```python
#ç•«å‡ºdecision tree
fig, ax = plt.subplots(figsize=(35, 20))
tree.plot_tree(dtree, fontsize=14)
plt.show()
```

### 2.8

```python
score50 = dtree.score(X_test, y_test)
score50
```

### 2.9

```python
#è©•ä¼°æ±ºç­–æ¨¹æ¨¡åž‹å¥½å£ž
pred = dtree.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,pred))
```

### 2.10

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=12,stratify=y)
train60 = model.fit(X_train, y_train)
score60 = train60.score(X_test, y_test)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=12,stratify=y)
train70 = model.fit(X_train, y_train)
score70 = train70.score(X_test, y_test)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12,stratify=y)
train80 = model.fit(X_train, y_train)
score80 = train80.score(X_test, y_test)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=12,stratify=y)
train90 = model.fit(X_train, y_train)
score90 = train90.score(X_test, y_test)

data = {
    'split_proportion':['50/50','60/40','70/30','80/20','90/10'],
    'score':[score50,score60,score70,score80,score90]
}
scoredf = pd.DataFrame(data)
scoredf.head()
```

### 2.11

```python
plt.plot(scoredf["split_proportion"], scoredf["score"], label = 'score')  

plt.xlabel("train/test proportion")
plt.title("train/test split proportion & result", fontsize = 15, y = 1)
plt.xticks(rotation=90)
plt.show()
```
